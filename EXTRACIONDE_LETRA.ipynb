```python
import pandas as pd
```

```python
import requests

genero='reggaeton'
response = requests.get('https://www.letras.com/mais-acessadas/'+genero+'/')
string_xml = response.content
```

```python
from bs4 import BeautifulSoup as Bs
```

```python
string_xml
```

```python
Soup= Bs(string_xml)
tags = Soup.find_all('ol' ,class_='top-list_mus cnt-list--col1-3')
```

#hasta aqui tenemos el listado de url para ir a ver la letra de la musica , y tambien tenemos el titulo


```python
Soup2= Bs(str(tags))
```

```python
tags2 = Soup2.find_all('a')
```

```python
raiz='https://www.letras.com'
url_letras_mu=[]

url_title_mu=[]
for tag in tags2:
  #sacamos el url completo donde esta la letra de la cancion
  url_letras_mu.append(raiz+tag.get('href'))
  #sacamos el titulo de la musica
  url_title_mu.append(tag.get('title'))
```

```python
from newspaper import Article
```

```python
t=''
```

```python
cantidad = 2
i=0
letras_mu=[]
while cantidad > i:
  Articulo=Article(url_letras_mu[i],language='es')
  Articulo.download()
  Articulo.parse()

  t = Articulo.text.replace("\n\n", " ")
  letras_mu.append(t)
  t=''
  i+=1
  
  
```

```python
letras_mu
```

```python
df = pd.DataFrame()
```

```python
df['link']=url_letras_mu[:cantidad]
```

```python
df['title']=url_title_mu[:cantidad]
```

```python
df['texto']=letras_mu
```

```python
#guarda el dataframe en un excel
writer = pd.ExcelWriter('letras_de_canciones.xlsx')
df.to_excel(writer, 'principal')
writer.save()
writer.close()
```
